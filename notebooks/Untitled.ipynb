# Texto: Cantos Fantasias

```python
import pandas as pd
```

```python
df = pd.read_csv('cantos_fantasias.csv')
df.head()
```

```python
df.shape
```

```python
lema__value_counts = df.lemma.value_counts()
lema__value_counts.head()
```

```python
lemma_value_counts_percent = df.lemma.value_counts(normalize=True)
lemma_value_counts_percent.head()
```

A distribuição de frequência da variável lemma não nos diz muita coisa. Talvez podemos melhorar esse cenário se excluirmos potuação.

```python
df__no_pontuation = df[df['is_punct'] == False]
df__no_pontuation_no_space = df__no_pontuation[df['tag'] != 'SPACE']
df__no_pontuation_no_space.head(40)
```

```python
df__no_pontuation_no_space.shape
```

```python
lema_distr_freq = df__no_pontuation_no_space.lemma.value_counts(normalize=True)
lema_distr_freq.head(10)
```

```python
pos_distr_freq = df__no_pontuation_no_space.pos.value_counts(normalize=True)
pos_distr_freq.head(10)
```

```python
dep_distr_freq = df__no_pontuation_no_space.dep.value_counts(normalize=True)
dep_distr_freq.head(10)
```
